### TODO #4: 利用 Seaborn 進行 Heatmap 作圖，並進行說明
- 觀察每一個特徵與房價 (MEDV) 的關聯程度
- 使用 dr.corr() 來計算 feature 和 feature 之間的關係係數

```python
plt.figure(figsize=(16, 9))
# Please use a function of Pandas here.
df_corr = df.corr()
# Plot a heatmap with Seaborn
sns.heatmap(df_corr, annot = True) # annot = True -> 可看見相關係數的數值
```

#### 說明
透過呼叫corr()可以得到每一個feature之間的相關係數，然後這邊用heatmap把他表示出來，越接近1則越白，反之越接近-1則黑。

因為我在plot heatmap的時候有設定annot = True，所以可以更清楚的看到在數值上的關聯性，以便去挑選那些feature和MEDV較相關。

### TODO #5: 挑選特徵，並進行說明
- 觀察 heatmap 中具有"顯著"正相關或是"顯著"負相關的特徵
- 把這些特徵進行 concatenation，得到新的 dataframe

```python
# RM 和 MEDV 呈現顯著正相關
# LSTAT 和 MEDV 呈現顯著負相關
X = pd.DataFrame(np.c_[df['RM'], df['LSTAT']], # Please put features inside the bracket
                 columns =  ['RM', 'LSTAT']# Please put feature names here 
                )
```

#### 說明
經由前面的heatmap，可以看出和MEDV最正相關的是RM這個feature，而最負相關的是LSTAT這個feature，因此選擇把這兩個特徵做concatenation。

### TODO #6: 呈現特徵分佈，並進行說明

```python
# Please plot subplots here
plt.figure(figsize=(16, 9)) # 調整圖片尺寸
y = df['MEDV'] # 把Y軸設成MDEV

plt.subplot(1, 2, 1) # subplot分為一列兩行，先取第一行
x = X['RM'] # X軸設成RM
plt.scatter(x, y) # 畫出 scatter plot
# 設定title
plt.title('RM')
plt.xlabel('RM')
plt.ylabel('MEDV')

plt.subplot(1, 2, 2) # subplot分為一列兩行，取第二行
x = X['LSTAT'] # X軸設成 LSTAT
plt.scatter(x, y) # 畫出 scatter plot
# 設定title
plt.title('LSTAT')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
```

## 訓練模型前的準備
### TODO #7: 切分資料
- 利用 train_test_split，將資料切成 train set 和 test set
- 設定 train set 的比例為全部資料的 80%
- 請記得設定 random_state 以控制隨機過程

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, df['MEDV'], test_size = 0.2, random_state = 325)

print(X_train.shape) #(404, 2)
print(X_test.shape) #(102, 2)
print(Y_train.shape) #(404, )
print(Y_test.shape) #(102, )
```

## 訓練模型並進行評估

### TODO #8: 訓練 Linear Regression 模型，並說明訓練結果
- 利用 scikit-learn，將資料 fit 在 LinearRegression
- 將結果繪製成散佈圖 (scatter plot)，其中 X軸 為 Y_pred、Y軸為 Y_true

```python
# Modeling
from sklearn.linear_model import LinearRegression

model_LinearRegression = LinearRegression() # modeling
model_LinearRegression.fit(X_train,Y_train) # 訓練模型
```

```python
# plotting the y_test vs y_pred
Y_pred = model_LinearRegression.predict(X_test) # 將預測結果放入Y_pred
Y_true = Y_test # Y_true為測試集的Y_test

plt.scatter(Y_pred, Y_true) # 畫出scatter plot
plt.xlabel('Y_pred')
plt.ylabel('Y_true')
LinearRegression_score = metrics.r2_score(Y_true, Y_pred) # R2 score
print('R2_LinearRegression:', LinearRegression_score) # R2 score
```

#### 說明
在這個scatter plot中，X軸是testing set，Y軸是predict的結果，因此點所表現出來的線就能代表做出來的結果和解答的相關程度。

這邊可以看出斜率大致也是顯著的正相關，代表訓練的結果是還不錯的。

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
import xgboost as xgb
from sklearn import metrics
from sklearn.metrics import mean_squared_error
```

```python
init_params = { 'n_estimators': 100,
                'max_depth': 3,
                'min_child_weight': 2,
                'subsample': 0.6,
                'colsample_bytree': 0.6,
                'Learning_rate': 0.05}
###########################################################

n_estimators = {'n_estimators': [100, 200, 300, 400, 500]}
init_XGBmodel = xgb.XGBClassifier(**init_params)
grid = GridSearchCV(init_XGBmodel, param_grid = n_estimators, refit = True, cv = 5, n_jobs = -1) # five-fold validation
grid.fit(X_train, Y_train)
best_parameters = grid.get_params()

###########################################################

max_depth = {'max_depth': [3, 4, 5, 6, 7]}
init_XGBmodel = xgb.XGBClassifier(**best_parameters)
grid = GridSearchCV(init_XGBmodel, param_grid = max_depth, refit = True, cv = 5, n_jobs = -1) # five-fold validation
grid.fit(X_train, Y_train)
best_parameters = grid.get_params()

###########################################################

min_child_weight = {'min_child_weight': [2, 3, 4, 5]}
init_XGBmodel = xgb.XGBClassifier(**best_parameters)
grid = GridSearchCV(init_XGBmodel, param_grid = min_child_weight, refit = True, cv = 5, n_jobs = -1) # five-fold validation
grid.fit(X_train, Y_train)
best_parameters = grid.get_params()

###########################################################

subsample = {'subsample': [0.6, 0.7, 0.8, 0.9]}
init_XGBmodel = xgb.XGBClassifier(**best_parameters)
grid = GridSearchCV(init_XGBmodel, param_grid = subsample, refit = True, cv = 5, n_jobs = -1) # five-fold validation
grid.fit(X_train, Y_train)
best_parameters = grid.get_params()

###########################################################

colsample_bytree = {'colsample_bytree': [0.6, 0.7, 0.8, 0.9]}
init_XGBmodel = xgb.XGBClassifier(**best_parameters)
grid = GridSearchCV(init_XGBmodel, param_grid = colsample_bytree, refit = True, cv = 5, n_jobs = -1) # five-fold validation
grid.fit(X_train, Y_train)
best_parameters = grid.get_params()

###########################################################

learning_rate = {'learning_rate': [0.05, 0.1, 0.15, 0.2]}
init_XGBmodel = xgb.XGBClassifier(**best_parameters)
grid = GridSearchCV(init_XGBmodel, param_grid = learning_rate, refit = True, cv = 5, n_jobs = -1) # five-fold validation
grid.fit(X_train, Y_train)
best_parameters = grid.get_params()
```

#### 說明
首先，為了讓grid search的結果更具有參考性，避免overfitting的問題，因此在做grid search時會用five-fold validation去做交叉驗證。

再來我的作法就是，一開始先初始化超參數，然後先選定一個超參數，例如我一開始是先選擇n_estimators，從這些數值當中用grid search去選擇一個表現最好的，然後把挑選出來的參數結果放進best_parameters裡面，接下來再換下一個超參數，做一樣的事情，但是這次會以上一次結果的best_parameters當作初始化來做，以此類堆，把其他的超參數都做一遍，用來得到最佳的組合。

其實一開始是想要一次去fit看看所有的組合，就是直接寫一個for loop然後把每一個超參數的組合都跑一遍來得到最佳解，但是後來發現這樣太花費時間，搞不好需要花到幾天的時間下去fit，所以最後選擇這樣的做法來縮減計算時間。
